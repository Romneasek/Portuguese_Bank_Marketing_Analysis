```python
# There are some binary columns (defualt, housing, loan) needed to convert intio numerical value.
# There are some categorical columns (job, marital, education, contact, month, day_of_week, and poutcome) needed to convert into numerical format.
# All feature columns are needed to convert into numerical values then only we can feed into the model.
```

```python
default_dummies = pd.get_dummies(data['default'], prefix = 'default')
default_dummies.tail()
```

```python
default_dummies.drop(['default_unknown'], axis=1, inplace=True)
default_dummies.tail()
```

```python
data = pd.concat([data, default_dummies], axis=1)
data.head()
```

```python
# Convert housing column into numerical value. 'yes' value to 1 and 'no' value to 0.
data['is_housing'] = data['housing'].apply(lambda row: 1 if row == 'yes' else 0)
```

```python
# View housing and is_housing columns together
data[['housing', 'is_housing']].head(10)
```

```python
# Convert loan column into numerical value. 'yes' value to 1 and 'no' value to 0.
data['is_loan'] = data['loan'].apply(lambda row: 1 if row == 'yes' else 0)
```

```python
# View loan and is_loan columns together
data[['loan', 'is_loan']].tail(10)
```

```python
# Creating one hot encoding for non-numeric marital column
# Convert categorical variable into dummy/indicator variables.
# For marital column, there are some values (married, single, divorced, D, M, S, and unkown)
```

```python
marital_dummies = pd.get_dummies(data['marital'], prefix = 'marital')
marital_dummies.tail()
```

```python
# Merge marital_dummies with marital column
pd.concat([data['marital'], marital_dummies], axis=1).head(n=10)
```

```python
# Removing marital_D, marital_M, marital_S, and marital_unknown
marital_dummies.drop(['marital_D', 'marital_M', 'marital_S', 'marital_unknown'], axis=1, inplace=True)
marital_dummies.head()
```

```python
# Merge marital_dummies into main dataframe
```

```python
data = pd.concat([data, marital_dummies], axis=1)
data.head()
```

```python
# Creating one hot encoding for job column
job_dummies = pd.get_dummies(data['job'], prefix = 'job')
job_dummies.tail()
```

```python
job_dummies.drop('job_unknown', axis=1, inplace=True)
```

```python
# Merge job_dummies into main dataframe
data = pd.concat([data, job_dummies], axis=1)
data.head()
```

```python
# Creating one hot encoding for education column
education_dummies = pd.get_dummies(data['education'], prefix = 'education')
education_dummies.head()
```

```python
education_dummies.drop(['education_unknown', 'education_basic-4y','education_basic-6y', 'education_basic-9y', 'education_illiterate'], axis=1, inplace=True)
education_dummies.tail()
```

```python
data = pd.concat([data, education_dummies], axis=1)
data.head()
```

```python
# Creating one hot encoding for contact column
contact_dummies = pd.get_dummies(data['contact'], prefix = 'contact')
contact_dummies.tail()
```

```python
data = pd.concat([data, contact_dummies], axis=1)
data.head()
```

```python
# Creating one hot encoding for poutcome column
poutcome_dummies = pd.get_dummies(data['poutcome'], prefix = 'poutcome')
poutcome_dummies.tail()
```

```python
data = pd.concat([data, poutcome_dummies], axis=1)
data.head()
```

```python
# Convert month column into numeric value
months = {'jan':1, 'feb':2, 'mar':3, 'apr':4, 'may':5, 'jun':6, 'jul':7, 'aug':8, 'sep':9, 'oct':10, 'nov':11, 'dec': 12}
data['month'] = data['month'].map(months)
data['month'].head()
```

```python
# Convert day_of_week column into numeric value
weeks = {'mon':1, 'tue':2, 'wed':3, 'thu':4, 'fri':5, 'sat':6, 'sun':7}
data['day_of_week'] = data['day_of_week'].map(weeks)
data['day_of_week'].head()
```

```python
# pdays column
# â€˜pdaysâ€™ column indicates the number of days that passed by after the client was last contacted from a previous campaign (numeric; 999 means client was not previously contacted). If the value of â€˜pdaysâ€™ is â€˜-1â€™, if so we will associate that with a value of 0
```

```python
data[data['pdays'] == -1]['pdays'].count()
```

```python
data['was_contacted'] = data['pdays'].apply(lambda row: 0 if row == -1 else 1)
data[['pdays','was_contacted']].head()
```

```python
data.head(77)
```

```python
data['age'] = data['age'].replace(to_replace=[ '35-30','30-35','25-30', '18-25', '30-32', '50-55', '45-50', '20-25', '20-40', '35-40', '30-40'], value=['33','33','27','22','31','53','47','23','30','38','35'])
```

```python
data['age'].unique()
```

```python
data['age'] = pd.to_numeric(data['age'])
```

```python
# Delete job, education, default, housing, loan, contact, month, pdays, poutcome columns
```

```python
data.drop(['job', 'education', 'marital', 'default', 'housing', 'loan', 'contact', 'pdays', 'poutcome'], axis=1, inplace=True)
```

```python
# View After converting all columns into numeric column
data.dtypes
```

```python
data.head(10)
```

```python
# Convert into X(features) and y(target)
X = data.drop('y', axis = 1)
y = data['y']
```

```python
X.shape
```

```python
y.shape
```

```python
# Splitting data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =32)
```

```python
# Print shape of data train and data test
print('Dimension of X_train dataframe:', X_train.shape)
print('Dimension of y_train dataframe:', y_train.shape)
print('Dimension of X_test dataframe:', X_test.shape)
print('Dimension of y_test dataframe:', y_test.shape)
```

```python
# Create a LogisticRegression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# Training the model
model.fit(X_train, y_train)
```

```python
# Predict test data
y_pred = model.predict(X_test)
```

```python
# Predicted and actual value
print("Predicted value: ", y_pred[:10])
print("Actual value: ", y_test[:10])
```

```python
# View accuracy of the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_pred = y_pred, y_true = y_test)
print(accuracy)
print(f'Accuracy of the model Logistic Regression is {accuracy*100:.2f}%')
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
rfcpredictions = rfc.predict(X_test)
print("Predicted value: ", rfcpredictions[:10])
print("Actual value: ", y_test[:10])
```

```python
accuracy = accuracy_score(y_pred = rfcpredictions, y_true = y_test)
print(f'Accuracy of the Random Forest Classifier model is {accuracy*100:.2f}%')
```

```python
# SVC (support Vector Classifier)
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
svcpredictions = svc.predict(X_test)
print("Predicted value: ", svcpredictions[:10])
print("Actual value: ", y_test[:10])
```

```python
accuracy = accuracy_score(y_pred = svcpredictions, y_true = y_test)
print(f'Accuracy of the SVC model is {accuracy*100:.2f}%')
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=0)
dtc.fit(X_train, y_train)
dtcprediction = dtc.predict(X_test)
print("Predicted value: ", dtcprediction[:10])
print("Actual value: ", y_test[:10])
```

```python
accuracy = accuracy_score(y_pred = dtcprediction, y_true = y_test)
print(f'Accuracy of the Decision Tree Classifier model is {accuracy*100:.2f}%')
```
